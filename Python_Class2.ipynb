### With Condition (Filter using if)

```python
evens = [i+4 for i in range(10) if i % 2 == 0]
print(evens)   # Output: [0, 2, 4, 6, 8]
```

##With if-else

```python
labels = ["Even" if i % 2 == 0 else "Odd" for i in range(5)]
print(labels)   # Output: ['Even', 'Odd', 'Even', 'Odd', 'Even']
```
