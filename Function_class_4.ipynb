###Function Introduction

```python
def greet(a , b):
  a = a**2
  b = b**2
  return a+b
sum = greet(10,20)
print(sum)
```

## Variabl size Argument


```python
def Squarer_addition(*args):
  #print(args)
  sum = 0
  for i in args:
    i = i**2
    sum = sum + i
  return sum



ans = Squarer_addition(1,2,3,4,5)
print(ans)
```

##Keyword Argument

```python
def student(**kwargs):
  for key, value in kwargs.items():
    print(f"{key} = {value}")
  #print(kwargs)

student(name = "Noman", clas = 10, roll = 30, marks = 30)
```

```python
def give_prediction():
  a = [10,20,30]
  b = [30,40,50]
  return a,b
x,y = give_prediction()
print(type(give_prediction()))
print(x)
print(y)
```

```python
lst = [x for x in range(500) ]

def data_loader(chunk_size,lst):
  for i in range(0,len(lst),chunk_size):
    yield lst[i:i+chunk_size]

x = data_loader(5,lst)

#print(type(x))
print(next(x))
print(next(x))
```
