##Practice Problem 1
Get the text from sample.txt, and count the following:
- Number of lines
- Number of words
- Number of characters

and save it to the counter of string.txt file

```python
from functools import reduce
##Readlines() function ta txt file er text guloke list akare show kore
with open("./sample_data/sample.txt","r") as file:
  string = file.readlines()


  #print(string)

  #line count
  totallines  = len(string)
  print(totallines)

  #word count
  number_of_words = list(map(lambda x:len(x.split()),string))
  print(number_of_words)

  total_No_of_words = reduce(lambda x , y : x + y,number_of_words)
  # print(total_No_of_words)

  string_list = list(map(lambda x: x.strip(),string))
  print(string_list)

  #Reduce space
  string_list = list(map(lambda x: x.replace(" ",""),string_list))
  print(string_list)
  print('\n')
  string_list = list(map(lambda x: len(x),string_list))
  print(string_list)
   #total count of character
  total_count_of_characters = reduce(lambda x , y : x + y,string_list)
  print(total_count_of_characters)

  with open("./sample_data/counter.txt","w") as file:
    file.write(f"tital No of words:{total_No_of_words}\ntotal No of counters:{total_count_of_characters}\ntotal Lines:{totallines}")
```
